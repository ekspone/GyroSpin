```python
import numpy as np
import matplotlib.pyplot as plt
from time import time
```

```python
from diffeqpy import de, ode
```

```python
def f(u,p,t):
    x, y, z = u
    sigma, rho, beta = p
    return [sigma * (y - x), x * (rho - z) - y, x * y - beta * z]

u0 = [1.0,0.0,0.0]
tspan = (0., 100.)
p = [10.0,28.0,8/3]
prob = de.ODEProblem(f, u0, tspan, p)
sol = de.solve(prob,saveat=0.01)

```

```python
plt.plot(sol.t,de.transpose(de.stack(sol.u)))
plt.show()

us = de.stack(sol.u)
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(us[0,:],us[1,:],us[2,:])
plt.show()
```

```python
from GyroSolver import*
from PhysicalQte import *

g = 9.81           # m/s^2
m = 0.375          # kg
J3 = 0.000055      # kg.m^2
J1 = 0.000055      # kg.m^2
h = 31.25 * 1e-3   # m
x0 =  0.1
f_exc = 0.265
Phi = np.pi

J1_ = J1 + m * (h)**2

params = [g, m, h, J1, J3 , x0, Phi, f_exc]
CI = [np.pi / 6, 0, 0, 0, 0, 2 * np.pi * 200]

p_psi0 = p_psi_exp(params, CI)

Gyro_Carac_Values(params, CI)

p_psi0
```

```python
def the_dd_Free(t, g, h, m, J3, the, phi, psi, the_d, phi_d, psi_d, p_psi, J1_, w, p, K):
    the_dd_free = (m*g*h - phi_d * p_psi)*de.sin(the) / J1_ + 0.5 * de.sin(2*the)*(phi_d**2)
    the_dd_forced = - K * de.sin(w*t+p - phi) * de.cos(the) / J1_
    the_dd = the_dd_free + the_dd_forced
    return the_dd


def phi_dd_Free(t, g, h, m, J3, the, phi, psi, the_d, phi_d, psi_d, p_psi, J1_, w, p, K):
    phi_dd_free = - (2*the_d*phi_d / de.tan(the)) + (p_psi*the_d)/(J1_ * de.sin(the)) 
    phi_dd_forced = K * de.cos(w*t+p - phi) / (J1_ * de.sin(the))
    phi_dd = phi_dd_free + phi_dd_forced
    return phi_dd


def psi_dd_Free(t, g, h, m, J3, the, phi, psi, the_d, phi_d, psi_d, p_psi, J1_, w, p, K):
    psi_dd_free = (1 + de.cos(the)**2)*the_d*phi_d/de.sin(the) - p_psi*the_d/(J1_ * de.tan(the))
    psi_dd_forced = - K * de.cos(w*t+p - phi) / (J1_ * de.tan(the))
    psi_dd = psi_dd_free + psi_dd_forced
    return psi_dd
```

```python


def f(u,p,t):
    the, the_d, phi, phi_d, psi, psi_d = u
    
    g, m, h, J1_, J3, p_psi, w, Phi, K = p
    
    return [
            the_d,
            the_dd_Free(t, g, h, m, J3, the, phi, psi, the_d, phi_d, psi_d, p_psi, J1_, w, Phi, K),
            phi_d,
            phi_dd_Free(t, g, h, m, J3, the, phi, psi, the_d, phi_d, psi_d, p_psi, J1_, w, Phi, K),
            psi_d,
            psi_dd_Free(t, g, h, m, J3, the, phi, psi, the_d, phi_d, psi_d, p_psi, J1_, w, Phi, K),
            ]

t1 = time()

u0 = [np.pi / 6, 0, 0, 0, 0, 2 * np.pi * 200]
tspan = (0., 600.)

f_exc = 0.265
w = 2 * np.pi * f_exc
K = m * h * x0 * (w**2)
K = 0
p = [g, m, h, J1_, J3, p_psi0, w, Phi, K]
prob = de.ODEProblem(f, u0, tspan, p)

fastprob = de.jit(prob)

#de.Tsit5()
#RadauIIA5

sol = de.solve(fastprob, de.Tsit5(), abstol = 1e-8, reltol = 1e-8, maxiters=np.inf)#, alg_hints = ['stiff'])
t2 = time()

print('Duration :',t2-t1)
```

```python
us = de.stack(sol.u)
```
