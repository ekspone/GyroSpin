```python
import numpy as np
import matplotlib.pyplot as plt
from Mvt_Equations import *
from time import time
```

```python
CI = [np.pi - 0.001, 0, 0, 0, 0, 2*np.pi*50]

#t = np.linspace(0, 200, 1000000)

g_ = 9.81
m_ = 0.375
J3_ = 0.000055
J1_ = 0.000055
h_ = 31.25 * 1e-3
x0_ = 5 * 1e-2

omega_f =  0.265 * 4
params_f = [g_, m_, h_, J1_, J3_, x0_, np.pi,  omega_f]
```

```python
dt1 = time()

t, the, phi, psi, the_d, phi_d, psi_d = Solve_Gyro_Forced_Fast(0, 10, CI, params_f, solving_method='RK23')
plt.figure()
plt.plot(t, the* 180 / np.pi)
plt.show()
dt2 = time()
print(dt2 - dt1)
```

```python
f1 = 1
step_f = 0.001
f2 = 1.2
x1 = 2.5
x2 = 7.5
step_x = 2.5
print(Frequency_array(f1, f2, step_f))
print('Temps estimé =', len(Frequency_array(f1, f2, step_f)) * 5 / 3 )
print(Amplitude_array(x1, x2, step_x))
```

```python
dt1 = time()

f_max, exc_freq, tab_theta, tab_theta_min, tab_t = Frequency_Sweep_MIN(f1, f2, step_f, params_f, tf=50, CI=CI)

dt2 = time()
print(dt2 - dt1)
```

```python
plt.figure()
plt.plot(exc_freq, np.cos(tab_theta_min * np.pi / 180))
plt.xlabel(r'$\omega \; (Hz)$')
plt.ylabel(r'$Z_{max} / h$')
plt.show()
```

```python
tab_z_min = np.cos(tab_theta_min * np.pi / 180)

def Puissance(x, a, A):
    return A * (x**a)

from scipy.optimize import curve_fit

tab_bool = exc_freq >= 1.09
#tab_bool = (exc_freq >= f_max + 0.005) & (exc_freq <= 1.14)

f_fit = exc_freq[tab_bool]
angle_fit = tab_theta_min[tab_bool]


tab_params_fit, pcov = curve_fit(Puissance, f_fit, angle_fit)
print(tab_params_fit)


plt.figure()
plt.scatter(f_fit, angle_fit, marker='.')
plt.plot(f_fit, Puissance(f_fit, *tab_params_fit))
plt.show()
```

```python
f_fit = exc_freq[tab_bool]
z_fit = -np.cos(tab_theta_min[tab_bool] * np.pi / 180)


tab_params_fit, pcov = curve_fit(Puissance, f_fit, z_fit)
print(tab_params_fit)


plt.figure()
plt.scatter(f_fit, z_fit, marker='.')
plt.plot(f_fit, Puissance(f_fit, *tab_params_fit))
plt.show()
```
