```python
CI = [np.pi / 24, 0, 0, 0, 0, 2*np.pi*200]

g = 9.81
m = 0.375
J3 = 0.000055
J1 = 0.000055
h = 31.25 * 1e-3
x0 = 10 * 1e-2

omega_f =  0.265
params = [g, m, h, J1, J3, x0, np.pi,  omega_f]

t, the, phi, psi, the_d, phi_d, psi_d = Solve_Gyro_Forced(0, 150, CI, params)
```

```python
plt.figure(figsize=[12, 4])
plt.subplot(1, 3, 1)
plt.plot(t, the * 180 / np.pi)
plt.xlabel(r'$\theta(t)$')
plt.subplot(1, 3, 2)
plt.plot(t, phi * 180 / np.pi)
plt.xlabel(r'$\phi(t)$')
plt.subplot(1, 3, 3)
plt.plot(t, psi * 180 / np.pi)
plt.xlabel(r'$\psi(t)$')
plt.show()

plt.figure(figsize=[12, 4])
plt.subplot(1, 3, 1)
plt.plot(t, the_d)
plt.xlabel(r'$\dot{\theta}(t)$')
plt.subplot(1, 3, 2)
plt.plot(t, phi_d)
plt.xlabel(r'$\dot{\phi}(t)$')
plt.subplot(1, 3, 3)
plt.plot(t, psi_d)
plt.xlabel(r'$\dot{\psi}(t)$')
plt.show()
```

```python
from scipy.optimize import curve_fit

def Lin(x, a, b):
    return a *x + b

def Fit_Lin(x, y):
    tab, pcov = curve_fit(Lin, x, y)
    return tab

print('Omega_L exp :', Fit_Lin(t, phi))
print('Omega L th :', m * g * h / (J3 * 200 * 2 * np.pi))
```

```python
J1_ = J1 + m*(h**2)
p_psi = J3 * (psi_d + phi_d * np.cos(the))
p_the = J1_ * the_d
p_phi = J1_ * ((np.sin(the))**2) * phi_d + np.cos(the) * p_psi
```

```python
plt.figure(figsize=[12, 4])
plt.subplot(1, 3, 1)
plt.plot(t, p_the / J1_)
plt.xlabel(r'$p_\theta(t) / J_0$')
plt.subplot(1, 3, 2)
plt.plot(t, p_phi/ J1_)
plt.xlabel(r'$p_\phi(t) / J_0$')
plt.subplot(1, 3, 3)
plt.plot(t, p_psi / J1_)
plt.xlabel(r'$p_\psi(t) / J_0$')
plt.show()
```

```python
plt.figure()
plt.plot(t, p_phi / J1_)
plt.plot(t, 155 * np.cos(0.02288245 * t ))
plt.show()

plt.figure()
plt.plot(t, J1_ * phi_d * np.sin(the)**2, 'r')
plt.plot(t, p_psi * np.cos(the), 'b')
plt.plot(t, p_phi, 'k')
plt.show()

plt.figure()
plt.plot(t, p_psi, 'r')
plt.plot(t, J3 * 2 * np.pi * 200 * np.ones(len(t)), 'b' )
plt.show()

print(np.mean(p_psi))
print(np.std(p_psi, ddof = 1), np.std(p_psi, ddof = 1) / np.mean(p_psi))
print(J3 * 2 * np.pi * 200 )
```

```python
print(2 * np.pi * np.mean(p_psi) / (m * h * (2*np.pi*0.265)**2) )
```

```python
plt.figure()
plt.plot(t[t<=125], the[t<=125] * 180 / np.pi)
plt.show()

print('Omega_Theta exp :', Fit_Lin(t[t <=125], the[t<=125]))
T_the = 2 * np.pi / Fit_Lin(t[t <=125], the[t<=125])[0]
print('T_the =', T_the)
```

```python
N = ((p_phi - np.cos(the) * p_psi)**2) / (2 * J1_ * ((np.sin(the))**2))
EXC = - m*h*x0 * ((2 * np.pi * omega_f)**2) * np.cos(2*np.pi*omega_f*t + params[-2])*np.sin(the)*np.sin(phi)

plt.figure()
plt.plot(t, N, label=r'$\mathcal{N}$')
plt.plot(t, p_the**2 / (2 * J1_), label=r'$p_\theta$')
#plt.plot(t, p_psi**2 / (2 * J3), label=r'$p_\psi$')
#plt.plot(t, m * g * np.cos(the), label=r'$E_{pp}$')
plt.plot(t, EXC, label='EXC')

plt.xlabel(r'$\mathcal{N}(t)$')
plt.legend()
plt.show()
```

```python
t_f, the_f, phi_f, psi_f, the_d_f, phi_d_f, psi_d_f = Solve_Gyro_Free(0, 150, CI, params)

plt.figure(figsize=[12, 4])
plt.subplot(1, 3, 1)
plt.plot(t, the_f * 180 / np.pi)
plt.xlabel(r'$\theta(t)$')
plt.subplot(1, 3, 2)
plt.plot(t, phi_f * 180 / np.pi)
plt.xlabel(r'$\phi(t)$')
plt.subplot(1, 3, 3)
plt.plot(t, psi_f * 180 / np.pi)
plt.xlabel(r'$\psi(t)$')
plt.show()

plt.figure(figsize=[12, 4])
plt.subplot(1, 3, 1)
plt.plot(t, the_d_f)
plt.xlabel(r'$\dot{\theta}(t)$')
plt.subplot(1, 3, 2)
plt.plot(t, phi_d_f)
plt.xlabel(r'$\dot{\phi}(t)$')
plt.subplot(1, 3, 3)
plt.plot(t, psi_d_f)
plt.xlabel(r'$\dot{\psi}(t)$')
plt.show()

p_psi_f = J3 * (psi_d_f + phi_d_f * np.cos(the_f))
p_the_f = J1_ * the_d_f
p_phi_f = J1_ * ((np.sin(the_f))**2) * phi_d_f + np.cos(the_f) * p_psi_f

plt.figure(figsize=[12, 4])
plt.subplot(1, 3, 1)
plt.plot(t, p_the_f / J1_)
plt.xlabel(r'$p_\theta(t) / J_0$')
plt.subplot(1, 3, 2)
plt.plot(t, p_phi_f/ J1_)
plt.xlabel(r'$p_\phi(t) / J_0$')
plt.subplot(1, 3, 3)
plt.plot(t, p_psi_f / J1_)
plt.xlabel(r'$p_\psi(t) / J_0$')
plt.show()

N_f = ((p_phi_f - np.cos(the_f) * p_psi_f)**2) / (2 * J1_ * ((np.sin(the_f))**2))

plt.figure()
plt.plot(t, N_f, label=r'$\mathcal{N}_f$')
plt.plot(t, p_the_f**2 / (2 * J1_), label=r'$p_\theta f$')
#plt.plot(t, p_psi_f**2 / (2 * J3), label=r'$p_\psi f$')
#plt.plot(t, m * g * np.cos(the_f), label=r'$E_{pp} f$')

plt.xlabel(r'$\mathcal{N}(t) FREE$')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[12, 8])
plt.subplot(2, 2, 1)
plt.plot(t, N, label=r'$FORCED$')
plt.plot(t, N_f, label=r'$FREE$')
plt.xlabel(r'$\mathcal{N}$')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(t, m * g * h *np.cos(the), label=r'$FORCED$')
plt.plot(t, m * g * h *np.cos(the_f), label=r'$FREE$')
plt.xlabel(r'$EPP$')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(t, p_psi**2 / (2 * J3), label=r'$FORCED$')
plt.plot(t, p_psi_f**2 / (2 * J3), label=r'$FREE$')
plt.xlabel(r'$p_\psi$')
plt.legend()


plt.subplot(2, 2, 4)
plt.plot(t, p_the**2 / (2 * J1_), label=r'$FORCED$')
plt.plot(t, p_the_f**2 / (2 * J1_), label=r'$FREE$')
plt.xlabel(r'$p_\theta$')
plt.legend()
plt.show()
```
