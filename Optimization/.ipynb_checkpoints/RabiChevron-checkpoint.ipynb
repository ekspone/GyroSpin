# Trying to make Rabi's Chevrons :

```python
tab_t_burst = np.linspace(0, 20, 101, endpoint=True)
tab_freq = np.linspace(0.2, 0.3, 101, endpoint=True)
print(tab_freq)
tab_freq = tab_freq * 4
print(len(tab_freq))
print('Estimated time (s) :', len(tab_freq) * 20 / 60)
```

```python
t1 = time()
tab_theta, list_the, list_t = Rabi_Chevron(tab_freq, tab_t_burst, params_f, CI)
t2 = time()
print(t2 - t1)
```

```python
plt.figure(figsize=[10,10])
plt.imshow(1-np.cos(tab_theta))
plt.colorbar()
plt.xticks(np.arange(0, 101, 5), tab_t_burst[np.arange(0, 101, 5)].astype(int) ) 
plt.yticks(np.arange(0, 101, 5), tab_freq[np.arange(0, 101, 5)].round(4) ) 
plt.xlabel(r'$t_\text{burst} \; (s)$')
plt.ylabel(r'$f_\text{exc} \; (s^{-1})$')
plt.show()
```
